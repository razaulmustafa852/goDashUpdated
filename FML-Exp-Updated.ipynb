<h3>Beggining , Middle and Last User Experience</h3>

```python
import numpy
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set_style("whitegrid")
```

```python
df = pd.read_csv("Percentile - Sheet1.csv")
```

```python
competition = [5, 10 , 15]
```

```python
#for c in range(2): 
 #   print(competition[c])
```

```python
delays = [10, 50, 100]
```

```python
nodesColors = ["#3f8a27", "#fac507", "#fa0707"]
```

```python
algos  = ['conv', 'arbiter', 'elastic', 'geometric', 'average','exponential', 'logistic', 'bba']
```

```python
for c in range(3): 
    for d in delays:
        
        for x in algos:
            if competition[c] == 5:
                aSel = df[(df['Competition'] == 5) & (df['Algorithms'] == x)]
                Nodes = aSel[(aSel['Hosts'] == 1) | (aSel['Hosts'] == 3) | (aSel['Hosts'] == 5)]
                #D100 = Nodes[(Nodes['Delay'] == delays[d])]
                #print('135')
            elif competition[c] == 10:
                aSel = df[(df['Competition'] == 10) & (df['Algorithms'] == x)]
                Nodes = aSel[(aSel['Hosts'] == 1) | (aSel['Hosts'] == 5) | (aSel['Hosts'] == 10)]
                #D100 = Nodes[(Nodes['Delay'] == delays[d])]
            elif competition[c] == 15:
                aSel = df[(df['Competition'] == 15) & (df['Algorithms'] == x)]
                Nodes = aSel[(aSel['Hosts'] == 1) | (aSel['Hosts'] == 10) | (aSel['Hosts'] == 15)]
                #D100 = Nodes[(Nodes['Delay'] == delays[d])]
        
            D100 = Nodes[(Nodes['Delay'] == d) ]
            g = sns.FacetGrid(D100, col="Bandwidth",hue="Hosts",palette=nodesColors, hue_kws={"marker": ["*", "o","x"]})
            g.map(plt.scatter, "MOS", "Percentile")
            g.set_axis_labels("P.1203", "Percentile")
            axes = g.axes.flatten()
            axes[0].set_title("5 Mbps")
            axes[1].set_title("10 Mbps")
            axes[2].set_title("15 Mbps")
            g.add_legend(title= 'Host #')
            plt.subplots_adjust(top=0.80)
            g.fig.suptitle(x.capitalize() + ' &  Delay '+str(d)+' ms, Nodes competitions '+ str(competition[c]) +'')
        print('delay '+ str(d) +' & competition '+ str(competition[c]) +' user experience on each algorithm')  
```
