```python
import numpy
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("QOEALL - Sheet1.csv")
```

```python
df.head()
```

```python
competition = [5]
```

```python
delays = [100]
```

```python
bandwidth = [5]
```

```python
nodesColors = ["#3f8a27", "#fac507", "#fa0707"]
```

```python
algos  = ['conv', 'arbiter', 'elastic', 'geometric', 'average','exponential', 'logistic', 'bba']
```

```python
for c in range(1):
    for b in bandwidth:
        for d in delays:
        
            for x in algos:
                if competition[c] == 5:
                    al = df[(df['Competitions'] == 5) & (df['Algo'] == x) & (df['Bandwidth'] == b)& (df['Delay'] == d)]
                    #Nodes = aSel[(aSel['User'] == 1)]
                    #D100 = Nodes[(Nodes['Delay'] == delays[d])]
                    #print('135')
                elif competition[c] == 10:
                    al = df[(df['Competitions'] == 10) & (df['Algo'] == x) & (df['Bandwidth'] == b)& (df['Delay'] == d) ]
                    #Nodes = aSel[(aSel['User'] == 1)]
                    #D100 = Nodes[(Nodes['Delay'] == delays[d])]
                elif competition[c] == 15:
                    al = df[(df['Competitions'] == 15) & (df['Algo'] == x) & (df['Bandwidth'] == b)& (df['Delay'] == d)]
                    #Nodes = aSel[(aSel['User'] == 1)]
                    #D100 = Nodes[(Nodes['Delay'] == delays[d])]
        
                aSel = al[(al['User'] == 1)]
                fig, axs = plt.subplots(6,figsize=(7,6))
                fig.suptitle(x.upper() + ' : Host # 1, Delay '+str(d)+' ms & Bandwidth '+ str(b) +' Mbps on competition '+str(competition[c])+'')
                axs[0].plot(aSel['IntSeg'],aSel['Resolutions'],'tab:purple')
                axs[0].set(ylabel='Resolutions')
                axs[1].plot(aSel['IntSeg'], aSel['P1203'],'tab:orange')
                axs[1].set(ylabel='P1203')
                axs[2].plot(aSel['IntSeg'],aSel['Clae'],'tab:green')
                axs[2].set(ylabel='Clae')
                axs[3].plot(aSel['IntSeg'],aSel['Duanmu'],'tab:red')
                axs[3].set(ylabel='Duanmu')
                axs[4].plot(aSel['IntSeg'],aSel['Yin'],'tab:blue')
                axs[4].set(ylabel='Yin')
                axs[5].plot(aSel['IntSeg'],aSel['Yu'],'tab:brown')
                axs[5].set(ylabel='Yu',xlabel='Segments')
                for ax in axs.flat:
                    ax.label_outer()
```
