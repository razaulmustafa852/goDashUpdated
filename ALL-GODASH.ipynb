<h2>Comparison of throughput based (Conventional) vs Buffered (Logistic) Algorithm <h2>

<h3>Conventional (Throughput Based Algorithm ) <h3>

```python
nodesColors = ["#3f8a27", "#fac507", "#fa0707"]
```

<h2>Delay 10 ms </h2>

```python
#CONV Throughput based
conv_D10 = df[(df['Algorithms'] == 'conv') & (df['Delay'] == 10)]
```

```python
g = sns.FacetGrid(conv_D10, col="Bandwidth", hue="Competition",palette=nodesColors, hue_kws={"marker": ["*", "o","x"]})
g.map(plt.scatter, "MOS", "Percentile")
g.add_legend(title= 'Competition')
g.set_axis_labels("P.1203", "Percentile")
axes = g.axes.flatten()
plt.subplots_adjust(top=0.90)
g.fig.suptitle('Conventional: Bandwidth 5, 10, 15 mbit/s & delay 10 ms')
axes[0].set_title("")
axes[1].set_title("")
axes[2].set_title("")
```

```python
g = sns.FacetGrid(conv_D10, col="Bandwidth", hue="Competition",palette=nodesColors, hue_kws={"marker": ["*", "o","x"]})
g.map(plt.scatter, "MOS", "ResolutionsV")
g.add_legend(title= 'Competition')
g.set_axis_labels("P.1203", "Resolutions")
axes = g.axes.flatten()
plt.subplots_adjust(top=0.90)
g.fig.suptitle('Conventional: Bandwidth 5, 10, 15 mbit/s & delay 10 ms')
axes[0].set_title("")
axes[1].set_title("")
axes[2].set_title("")
```
